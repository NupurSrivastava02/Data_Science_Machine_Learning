```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

```python
x = np.array([1,2,3,4,5])
```

```python
x
```

```python
x = x.reshape(5,1)
```

```python
x.shape
```

```python
y = np.array([200,300,400,500,600])
```

```python
y = y.reshape(5,1)
```

```python
y.shape
```

```python
plt.scatter(x,y)
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='purple')
plt.xlabel('Independent')
plt.ylabel('Dependent')
plt.show()
```

```python
#Lasso process
```

```python
from sklearn.linear_model import Lasso
```

```python
model2 = Lasso(alpha=20)
```

```python
model2.fit(x,y)
```

```python
model2.score(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,model2.predict(x), color='pink')
plt.xlabel('Independent')
plt.ylabel('Dependent')
plt.show()
```

```python
# Ridge Process
```

```python
from sklearn.linear_model import Ridge
```

```python
model3 = Ridge(alpha=1)
```

```python
model3.fit(x,y)
```

```python
model3.score(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x, model3.predict(x), color = 'yellow')
plt.xlabel('Independent')
plt.ylabel('Dependent')
plt.show()
```
